```python
import matplotlib.pyplot as plt
activities = ['Sleep', 'Study', 'Exercise', 'Entertainment']
proportions = [8, 6, 2, 8]

plt.figure(figsize=(8, 6))
plt.pie(proportions, labels=activities, autopct='%1.1f%%')
plt.title('Daily Activity Distribution')

plt.show()
```
