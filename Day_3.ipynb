```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
# Create a plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Prime Numbers')
# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Line Plot')
plt.legend()
# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

name = ['lola', 'hola', 'nola', 'mola']
income = [20, 40, 35, 50]
plt.bar(name, income, color='skyblue')

plt.xlabel('Names')
plt.ylabel('Paisa')
plt.title('Bar Chart Example')

plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.scatter(x, y, color='Blue', label='Data Points',marker = "*")

plt.xlabel('paradyes')
plt.ylabel('Pink')
plt.title('Monthly sales')
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Harshala', 'Prachi', 'Abhay', 'Nikita']
sizes = [15, 30, 45, 10]
colors = ['silver', 'green', 'pink', 'lightgreen']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.f%%', startangle=140)

plt.title('Friends')

plt.show()
```
